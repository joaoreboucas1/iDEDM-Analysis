# Plotting LSST-Y1 Chains

```python
import numpy as np
import matplotlib.pyplot as plt
import getdist
from getdist import plots as p
```

```python
codes = [20, 21, 22, 23, 24, 25, 26, 27]
chains = {}
for code in codes:
    samples = getdist.loadMCSamples(f"./chains/lsst_y1/MCMC{code}")
    samples.removeBurn(0.55)
    samples.addDerived(samples["sigma8"] * np.sqrt(samples["omegam"]/0.3), name='S8', label="S_8")
    chains[code] = samples
```

```python
plot = p.get_subplot_plotter()
plot.settings.rc_sizes(axes_fontsize=20, lab_fontsize=20, legend_fontsize=25)
plot.triangle_plot(
    [chains[21], chains[20]],
    ['omegam', 'sigma8', 'S8', 'xi_interaction'],
    filled=True, 
    param_limits={'omegam': [0.25, 0.35], 'S8': [0.81, 0.91], 'sigma8': [0.78, 1.0], 'xi_interaction': [-0.1, -0.01]},
    legend_labels=[r"$\xi \Lambda$CDM", r"$\Lambda$CDM"],
    legend_loc='upper right',
	prop={"size": "15"})
plt.suptitle(f"LSST-Y1 Analyses with $\Lambda$CDM Fiducial DV", y=1.03)
plt.savefig(f"Figures/LSST_3x2_NO_INTERACTION_FIDUCIAL.pdf", bbox_inches="tight")
```

```python
plot = p.get_subplot_plotter()
plot.settings.rc_sizes(axes_fontsize=20, lab_fontsize=20, legend_fontsize=25)
plot.triangle_plot(
    [chains[23], chains[22]],
    ['omegam', 'sigma8', 'S8', 'xi_interaction'],
    filled=True, 
    param_limits={'S8': [1.0, 1.3], 'sigma8': [0.8, 1.3], 'xi_interaction': [-0.6, -0.2]},
    legend_labels=[r"$\xi \Lambda$CDM", r"$\Lambda$CDM"],
    legend_loc='upper right')
plt.suptitle(r"LSST-Y1 Analyses with $\xi \Lambda$CDM Fiducial DV", y=1.04)
plt.savefig("Figures/LSST_3x2_WITH_INTERACTION_FIDUCIAL.pdf", bbox_inches="tight")
```
