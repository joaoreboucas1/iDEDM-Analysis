# Literature Reproductions

```python
import getdist
import numpy as np
from getdist import plots
```

```python
# Loading Chains
analysis_settings = {u'ignore_rows':0.5}

chains = []
for i in range(1,7):
    chains.append(getdist.loadMCSamples('./chains/ACT_Exploration/MCMC{:d}'.format(i), settings=analysis_settings))
```

```python
for chain in chains:
    chain.addDerived(chain['sigma8'] * np.sqrt(chain['omegam']/0.3), name='S8', label='S_8')
```

```python
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot(chains[:3], ['omch2', 'xi_interaction', 'H0', 'omegam'], filled=True,
                #colors=['tab:red', 'tab:green', 'tab:blue'],
                #diag1d_kwargs={'lws':[2,2,2], 'colors':['tab:red', 'tab:green', 'tab:blue']},
                legend_labels=['ACT', 'ACT + BAO + Pantheon', 'ACT + BAO + Pantheon + CMB Lensing']
               )

g.export('Figures/ACT_EXPLORATION.pdf')
```

```python
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot([chains[0], chains[3]], ['omch2', 'xi_interaction', 'H0', 'omegam', 'S8'], filled=True,
                legend_labels=['ACT [Interacting]', 'ACT +[LCDM]']
               )
g.export('Figures/ACT_EXPLORATION_2.pdf')
```

```python
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot([chains[1], chains[4]], ['omch2', 'xi_interaction', 'H0', 'omegam', 'S8'], filled=True,
                legend_labels=['ACT + BAO + SNe [Interacting]', 'ACT + BAO + SNe [LCDM]']
               )
g.export('Figures/ACT_EXPLORATION_3.pdf')
```
