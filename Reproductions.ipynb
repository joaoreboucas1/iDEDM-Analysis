# Literature Reproductions

```python
import getdist
from getdist import plots
```

```python
# Loading Chains
analysis_settings = {u'ignore_rows':0.7}

chains = []
for i in range(1,4):
    chains.append(getdist.loadMCSamples('./chains/TEST{:d}'.format(i), settings=analysis_settings))
```

```python
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot(chains, ['omch2', 'xi_interaction', 'H0', 'omegam'], filled=True,
                #colors=['tab:red', 'tab:green', 'tab:blue'],
                #diag1d_kwargs={'lws':[2,2,2], 'colors':['tab:red', 'tab:green', 'tab:blue']},
                legend_labels=['Planck', 'Planck + BAO + Pantheon', 'Planck + R19']
               )
g.add_x_bands(74, 1, ax=(1,0), color='orange', alpha1=0.4, alpha2=0.2)

g.export('Figures/FIG3_EDV_1910_09853.pdf')
```

```python
chains2 = [chains[3], chains[1], chains[2]]
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot(chains2, ['H0', 'xi_interaction'], filled=True,
                colors=['tab:red', 'tab:green', 'tab:blue'], diag1d_kwargs={'lws':[2,2,2], 'colors':['tab:red', 'tab:green', 'tab:blue']},
                legend_labels=['Planck (sampled in h)', 'Planck + BAO + Pantheon', 'Planck + R19']
               )
g.add_x_bands(74, 1, ax=(1,0), color='orange', alpha1=0.4, alpha2=0.2)

g.export('Figures/FIG1_MATTEO_HOOPER_2020_second_try.pdf')
```
