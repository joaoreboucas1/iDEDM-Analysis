# Literature Reproductions

```python
import numpy as np
import getdist
from getdist import plots
```

```python
# Reproducing Figure 3 of 1910.09853
analysis_settings = {u'ignore_rows':0.4}

chains = []
for i in range(1,4):
    chains.append(getdist.loadMCSamples(f'./chains/test_chains_planck/TEST{i}', settings=analysis_settings))

g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot(chains, ['omch2', 'xi_interaction', 'H0', 'omegam'], filled=True,
                #colors=['tab:red', 'tab:green', 'tab:blue'],
                #diag1d_kwargs={'lws':[2,2,2], 'colors':['tab:red', 'tab:green', 'tab:blue']},
                legend_labels=['Planck', 'Planck + BAO + Pantheon', 'Planck + R19']
               )
g.add_x_bands(74, 1, ax=(1,0), color='orange', alpha1=0.4, alpha2=0.2)

g.export('Figures/FIG3_EDV_1910_09853.pdf')
```

```python
chains = {}
for code in [100, 101, 102, 103, 104, 105, 106]:
	chain = getdist.loadMCSamples(f"./chains/maglim_new/MCMC{code}")
	chain.removeBurn(0.6)
	chain.addDerived(chain['sigma8'] * np.sqrt(chain['omegam']/0.3), name="S8", label="S_8")
	chains[code] = chain
```

```python
g = plots.get_subplot_plotter(width_inch=4, subplot_size=3, subplot_size_ratio=8/10)
g.settings.scaling = False # prevent scaling down font sizes even though small subplots
g.settings.figure_legend_frame = False
g.rectangle_plot(
	['omegam'],
	['sigma8', 'S8'],
	roots=[chains[100], chains[101], chains[102]],
	nx=1,
    filled=[True, False, False],
    contour_ls=["-", "-", "-"],
	contour_lws=[2, 2, 2],
    colors=['gray', 'tab:blue', 'tab:orange'],
    legend_labels=['Fid. 3x2pt', r'$\xi_{\pm}$', r'$\gamma_t + w$'],
    lims=[0.16, 0.46, 0.680, 0.880],
	sharex=True
)

g.get_axes(0,0).set_xticks(np.linspace(0.2, 0.45, 6))
g.get_axes(0,0).set_yticks(np.linspace(0.6, 1, 9))
g.get_axes(1,0).set_yticks(np.linspace(0.7, 0.875, 8))
g.get_axes(0,0).set_ylim([0.56, 1])
g.get_axes(0,0).set_xlim([0.17, 0.46])
g.get_axes(1,0).set_ylim([0.68, 0.88])
g.export('Figures/Y3_Maglim_repr.pdf')
```
