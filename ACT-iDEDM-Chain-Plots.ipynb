```python
# GENERAL PLOT OPTIONS
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams['mathtext.rm'] = 'Bitstream Vera Sans'
matplotlib.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic'
matplotlib.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold'
matplotlib.rcParams['xtick.bottom'] = True
matplotlib.rcParams['xtick.top'] = False
matplotlib.rcParams['ytick.right'] = False
matplotlib.rcParams['axes.edgecolor'] = 'black'
matplotlib.rcParams['axes.linewidth'] = '1.0'
matplotlib.rcParams['axes.labelsize'] = 'large'
matplotlib.rcParams['axes.grid'] = True
matplotlib.rcParams['grid.linewidth'] = '0.0'
matplotlib.rcParams['grid.alpha'] = '0.18'
matplotlib.rcParams['grid.color'] = 'lightgray'
matplotlib.rcParams['legend.labelspacing'] = 0.77
matplotlib.rcParams['savefig.bbox'] = 'tight'
matplotlib.rcParams['savefig.format'] = 'pdf'
matplotlib.rcParams['font.size'] = 25
```

```python
# Loading Chains
analysis_settings = {u'ignore_rows':0.5}
codes = [10*i + j for i in range(6) for j in range(1,7)]
labels = [
    r'CMB ($\xi$ $\Lambda$CDM)',
    r'CMB+BAO+SNe ($\xi$ $\Lambda$CDM)',
    r'CMB+BAO+SNe+CMBL ($\xi$ $\Lambda$CDM)',
    r'CMB ($\Lambda$CDM)',
    r'CMB+BAO+SNe ($\Lambda$CDM)',
    r'CMB+BAO+SNe+CMBL ($\Lambda$CDM)',
    r'CMB ($\xi$ qCDM)',
    r'CMB+BAO+SNe ($\xi$ qCDM)',
    r'CMB+BAO+SNe+CMBL ($\xi$ qCDM)',
    r'CMB (qCDM)',
    r'CMB+BAO+SNe (qCDM)',
    r'CMB+BAO+SNe+CMBL (qCDM)',
    r'CMB ($\xi$ pCDM)',
    r'CMB+BAO+SNe ($\xi$ pCDM)',
    r'CMB+BAO+SNe+CMBL ($\xi$ pCDM)',
    r'CMB (pCDM)',
    r'CMB+BAO+SNe (pCDM)',
    r'CMB+BAO+SNe+CMBL (pCDM)',
    r'CMB+DES ($\xi$ $\Lambda$CDM)',
    r'CMB+BAO+SNe+DES ($\xi$ $\Lambda$CDM)',
    r'CMB+BAO+SNe+CMBL+DES ($\xi$ $\Lambda$CDM)',
    r'CMB+DES ($\Lambda$CDM)',
    r'CMB+BAO+SNe+DES ($\Lambda$CDM)',
    r'CMB+BAO+SNe+CMBL+DES ($\Lambda$CDM)',
    r'CMB+DES ($\xi$ qCDM)',
    r'CMB+BAO+SNe+DES ($\xi$ qCDM)',
    r'CMB+BAO+SNe+CMBL+DES ($\xi$ qCDM)',
    r'CMB+DES (qCDM)',
    r'CMB+BAO+SNe+DES (qCDM)',
    r'CMB+BAO+SNe+CMBL+DES (qCDM)',
    r'CMB+DES ($\xi$ pCDM)',
    r'CMB+BAO+SNe+DES ($\xi$ pCDM)',
    r'CMB+BAO+SNe+CMBL+DES ($\xi$ pCDM)',
    r'CMB+DES (pCDM)',
    r'CMB+BAO+SNe+DES (pCDM)',
    r'CMB+BAO+SNe+CMBL+DES (pCDM)'
]
chains = {}
chain_labels = {}
for code, label in zip(codes, labels):
    chains[code] = getdist.loadMCSamples('./chains/ACT_Exploration/MCMC{:d}'.format(code), settings=analysis_settings)
    chain_labels[code] = label
```

```python
for code, chain in chains.items():
    chain.addDerived(chain['sigma8'] * np.sqrt(chain['omegam']/0.3), name='S8', label='S_8')
```

### Plotting all ACT $\xi \Lambda \mathrm{CDM}$ chains

```python
plot = plots.get_single_plotter(width_inch=10, ratio=1)
plot.settings.rc_sizes(axes_fontsize=27, lab_fontsize=29, legend_fontsize=24)
plot.triangle_plot([chains[1], chains[2], chains[3]],
                   ['omch2', 'xi_interaction', 'H0', 'omegam'],
                   filled=True,
                   contour_colors=['tab:red', 'tab:green', 'tab:blue'],
                   contour_lws=3,
                   legend_labels=[chain_labels[1], chain_labels[2], chain_labels[3]],
                   legend_loc='upper right')
plot.export('Figures/ACT_xiLCDM.pdf')
```
